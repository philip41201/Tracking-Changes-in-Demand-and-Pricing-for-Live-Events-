```python
import requests
import json
import pandas as pd
from sqlalchemy import create_engine
```

```python
api_url = 'https://app.ticketmaster.com/discovery/v2/events.json'
params = {'apikey':'API_KEY', 'keyword':'Garth Brooks', 'countryCode':'US'}
api_request = requests.get(api_url, params=params)
```

```python
details = json.loads(api_request.text)['_embedded']['events']
```

```python
event_details = {
    'event_id':[],
    'name':[],
    'genre':[],
    'venue':[],
    'city':[],
    'state':[],
    'publicOnSaleDate':[],
    'eventDate':[],
    'minPrice':[],
    'maxPrice':[],
}

for event in details:
    event_id = event['id']
    event_details['event_id'].append(event_id)
    print(event_id)
    
    name = event['name']
    event_details['name'].append(name)
    print(name)
    
    genre = event['classifications'][0]['genre']['name']
    event_details['genre'].append(genre)
    print(genre)
    
    venue = event['_embedded']['venues'][0]['name']
    event_details['venue'].append(venue)
    print(venue)
    
    city = event['_embedded']['venues'][0]['city']['name']
    event_details['city'].append(city)
    print(city)
    
    state = event['_embedded']['venues'][0]['state']['name']
    event_details['state'].append(state)
    print(state)
    
    publicOnSaleDate = event['sales']['public']['startDateTime'][0:10]
    event_details['publicOnSaleDate'].append(publicOnSaleDate)
    print(publicOnSaleDate)
    
    eventDate = event['dates']['start']['localDate']
    event_details['eventDate'].append(eventDate)
    print(eventDate)

    minPrice = event['priceRanges'][0]['min']
    event_details['minPrice'].append(minPrice)
    print(minPrice)
    
    maxPrice = event['priceRanges'][0]['max']
    event_details['maxPrice'].append(maxPrice)
    print(maxPrice)
 
'''
event_details['event_id'].append(details[6]['id'])
event_details['name'].append(details[6]['name'])
event_details['genre'].append(details[6]['classifications'][0]['genre']['name'])
event_details['venue'].append(details[6]['_embedded']['venues'][0]['name'])
event_details['city'].append(details[6]['_embedded']['venues'][0]['city']['name'])
event_details['state'].append(details[6]['_embedded']['venues'][0]['state']['name'])
event_details['publicOnSaleDate'].append(details[6]['sales']['public']['startDateTime'][0:10])
event_details['eventDate'].append(details[6]['dates']['start']['localDate'])
event_details['minPrice'].append(details[6]['priceRanges'][0]['min'])
event_details['maxPrice'].append(details[6]['priceRanges'][0]['max'])
'''
```
